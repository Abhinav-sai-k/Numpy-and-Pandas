#Facet Grid

It visualize multiple graphs




```python
df.head()
```

```python
sns.FacetGrid(data=df,row='Genre',col='Year',hue='Genre')
```

```python
fplot = sns.FacetGrid(data=df,row='Genre',col='Year',hue='Genre') #(Gives all possible subplots)
fplot= fplot.map(plt.scatter,'AudienceRating','CriticRating')  #(what to do in that subplot? pass plottype as functioncall and x,y arguments)
```

```python
fplot = sns.FacetGrid(data=df,row='Genre',col='Year',hue='Genre') #(Gives all possible subplots)

fplot.map(sns.histplot,'BudgetMillions')
```
