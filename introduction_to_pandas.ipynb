# Data **visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = (6,2)
import warnings
warnings.filterwarnings('ignore')
```

```python
plt1 = sns.distplot(df.InternetUsers)  #Dist plot is distributions plot(In this case uni variate analysis)
```

```python
plt2 = sns.distplot(df.InternetUsers,bins = 25)
```

```python
plt3 =  sns.displot(df.InternetUsers)  #Dis plot is nothing but histogram
```

```python
plt4 =  sns.boxplot(data = df,x = 'IncomeGroup',y = 'BirthRate')  #Box plot # Bi-variate Analysis!
```

```python
plt5  =sns.lmplot(data = df, x= 'InternetUsers',y = 'BirthRate')
```

```python
plt5  =sns.lmplot(data = df, x= 'InternetUsers',y = 'BirthRate',fit_reg= False)  # To remove regression line
```

```python
plt5  =sns.lmplot(data = df, x= 'InternetUsers',y = 'BirthRate',hue='IncomeGroup')
```
