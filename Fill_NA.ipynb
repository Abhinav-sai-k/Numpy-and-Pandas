# Filling NA values

```python
df
```

```python
df.info()
```

```python
import numpy as np
```

```python
from ast import match_case
var = df.isna().sum()>0
for i in df.columns:
  # print(i)
  if var[i]:
    print(f"Missing values in {i}:")
    print(df[i])
    print()
    print(df[i].dtype)
    print()
    match df[i].dtype:
      case np.float64:
        df[i].fillna(np.mean(df[i]),inplace=True)
      case np.object_:
        df[i].fillna(df[i].mode()[0],inplace=True)
      case _:
        print("Unknown Datatype Encountered")
```
